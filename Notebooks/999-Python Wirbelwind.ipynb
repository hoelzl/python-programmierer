### Mehrere Parameter, Default Argumente

```python
def add(a, b=0, c=0):
    return a + b + c
```

```python
print(add(2))
print(add(2, 3))
print(add(2, 3, 4))
print(add(1, c=3))
```

## Kontrollstrukturen

```python
def print_size(n):
    if n < 10:
        print("Very small")
    elif n < 15:
        print("Pretty small")
    elif n < 30:
        print("Average")
    else:
        print("Large")
```

```python
print_size(1)
print_size(10)
print_size(20)
print_size(100)
```

```python
for char in "abc":
    print(char, end='|')
```

```python
result = 0
for n in [1, 2, 3, 4]:
    result += n
result
```

## Dictionaries

```python
translations = {'snake': 'Schlange', 'bat': 'Fledermaus', 'horse': 'Hose'}
```

```python
print(translations['snake'])
print(translations.get('bat', '<unbekannt>'))
print(translations.get('monkey', '<unbekannt>'))
```

```python
# Fehler:
# translations['monkey']
```

```python
translations['horse'] = 'Pferd'
translations['horse']
```

```python
list = [1, 2, 3]
list[0] = 'one'
list
```

```python
for key in translations.keys():
    print(key, end=" ")
```

```python
for val in translations.values():
    print(val, end=" ")
```

```python
for item in translations.items():
    print(item, end=" ")
```

```python
for key, val in translations.items():
    print("Key:", key, "\tValue:", val)
```
